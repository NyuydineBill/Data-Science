```python
import cv2
from ultralytics import YOLO

# Loadx a pre-trained YOLOv11 model
model = YOLO("yolo11n.pt")  # Options: yolo11n.pt, yolo11s.pt, yolo11m.pt, yolo11l.pt, yolo11x.pt
```

# Making Predictions on an Image


```python
import matplotlib.pyplot as plt
```

```python
# Read an image
image = cv2.imread("coding.jpeg")

# Perform object detection
results = model.predict(image)

# Extract and display accuracy (confidence scores)
for result in results:
    for box in result.boxes:  # Iterate through detected bounding boxes
        confidence = box.conf.item()  # Convert confidence tensor to a scalar
        class_id = box.cls.item()  # Convert class ID tensor to a scalar
        print(f"Class ID: {class_id}, Confidence: {confidence:.2f}")

# Display results
for result in results:
    # Assuming result contains bounding boxes, class labels, and confidence scores
    result.plot()  # This will plot the results directly if supported by the library

# Alternatively, visualize using OpenCV or Matplotlib
# Convert BGR image to RGB for displaying with Matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(image_rgb)
plt.axis("off")
plt.show()
```
